LEVEL 1 (BASIC)
===============
Show all employees whose salary is NULL

Show all orders where discount is NOT NULL

Get products where category is NULL

Count number of employees with NULL manager_id


```sql
select * from EmployeesN
where salary is null
```

```sql

select * from orders
where discount is not null
```

```sql
select * from Products
where category is null
```

```sql
select count(*) as count_emp 
from EmployeesN
where manager_id is null
```

LEVEL 2 (ISNULL)
=================
Replace NULL salary with 0

Replace NULL bonus with 1000

Show order amount, if NULL replace with 500

Replace NULL stock with 0


```sql
select IFNULL(salary, 0) as updated_salary
 from EmployeesN
```

```sql
select ifnull(bonus, 1000) as replaced_bonus
from EmployeesN
```

```sql
select ifnull(amount, 500) as replaced_amount
from orders
```


LEVEL 3 (COALESCE)
===================

Show employee earnings using:
 salary, if NULL use bonus


```sql
select coalesce(salary, bonus) as updated
from EmployeesN
```

Show first available value:
 salary → bonus → 0


```sql
select coalesce(salary,bonus,0) as updated
from EmployeesN
```

Show product price:
 price → 1000 (default)


```sql
select coalesce(price,1000) as updated
from products
```

Get customer payment:
 amount → discount → 0


```sql
select coalesce(amount, discount,0) as updated
from orders
```

LEVEL 4 (NULLIF)
================
Convert salary to NULL if salary = 0

Convert discount to NULL if discount = 0


```sql
select nullif(salary,0) as updatedsalry
from EmployeesN
```

```sql
select nullif(discount,0) as updated_discount
from orders
```

LEVEL 5 (REAL-TIME SCENARIOS)
=============================
Calculate total earnings:

   salary + bonus (handle NULL properly)


```sql
select ifnull(salary,0) + ifnull(bonus,0) as total
from EmployeesN
```
